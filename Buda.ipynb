```python
import pandas as pd
import numpy as np
import requests
from datetime import datetime
```

```python
def getDf(first_timestamp, timestamp):
  market_id = 'btc-clp'
  url = f'https://www.buda.com/api/v2/markets/{market_id}/trades'
  entries = []
  flag = True
  while flag:
    response = requests.get(url, params={
      'timestamp': timestamp,
      'limit': 25
    }).json()
    entries = entries + response['trades']['entries']
    if (int(entries[-1][0]) < int(first_timestamp)):
      flag = False
    else:
      timestamp = entries[-1][0]
  data = np.array(entries)
  df = pd.DataFrame(data=data[0:, 0:-1], index=data[:,-1], columns=['timestamp', 'amount', 'price', 'direction'])
  df = df.query('timestamp >= @first_timestamp')
  df['traded'] = df['amount'].astype(float) * df['price'].astype(float)
  return df
```

### 1) ¿Cuánto dinero (en CLP) se transó durante el evento "Black Buda" BTC-CLP ? (truncar en 2 decimales)
Answer: 35583121.00

```python
first_timestamp = '1709294400000' # 2024-03-01 12:00:00
timestamp = '1709298000000' # 2024-03-01 13:00:00

df = getDf(first_timestamp, timestamp)
df
```

```python
traded_amount = df['traded'].astype(float).sum()
traded_amount = traded_amount
traded_amount
```

### 2) En comparación con el mismo día del año anterior, ¿cuál fue el aumento porcentual en el volumen de transacciones (en BTC)? (truncar en 2 decimales)
Answer: 176.42

```python
first_timestamp_ly = '1677672000000' # 2024-03-01 12:00:00
timestamp_ly = '1677675600000' # 2024-03-01 13:00:00

df_ly = getDf(first_timestamp_ly, timestamp_ly)
df_ly
```

```python
traded_amount_ly = df_ly['traded'].astype(float).sum()
traded_amount_ly = traded_amount_ly
traded_amount_ly
```

```python
percentage = ((traded_amount - traded_amount_ly) / traded_amount_ly) * 100
percentage = percentage
percentage
```
